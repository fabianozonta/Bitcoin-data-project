```python
import yfinance as yf
import datetime

# Defina a data de início de acordo com os dados disponíveis
start_date = '2015-01-01'
end_date = '2023-11-20'

# Obtenha os dados do Yahoo Finance usando yfinance
bitcoin = yf.Ticker('BTC-USD')
bitcoin_data = bitcoin.history(start=start_date, end=end_date).reset_index()

# Exiba as primeiras linhas do DataFrame
print(bitcoin_data.head())
```

```python
# exemplo simplificado usando Scikit-learn com um modelo de regressão linear:

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Supondo que 'X' seja o tempo e 'y' seja o preço do Bitcoin
X = np.arange(len(bitcoin)).reshape(-1, 1)
y = bitcoin['Close']

# Dividir os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.6, shuffle=False)

# Criar um modelo de regressão linear
model = LinearRegression()

# Ajustar o modelo aos dados de treino
model.fit(X_train, y_train)

# Prever os dados de teste
y_pred = model.predict(X_test)

# Avaliar o desempenho do modelo
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
# modelo de séries temporais

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Supondo que 'bitcoin_data' seja o DataFrame que você criou anteriormente

# Ajustar o modelo ARIMA
order = (1, 1, 1)  # Parâmetros p, d, q do ARIMA
model = ARIMA(bitcoin_data['Close'], order=order)
result = model.fit()

# Resumo do modelo
print(result.summary())

# Visualizar os resíduos do modelo
residuals = result.resid
plt.figure(figsize=(12, 6))
plt.plot(residuals)
plt.title('Resíduos do Modelo ARIMA')
plt.show()

# Fazer previsões até o final de 2024
forecast_end_date = pd.to_datetime('2024-12-31').tz_localize('UTC')  # Configurar fuso horário para UTC
forecast_steps = (forecast_end_date - bitcoin_data.index[-1]).days
forecast = result.get_forecast(steps=forecast_steps)

# Extrair os valores da previsão e intervalo de confiança
forecast_values = forecast.predicted_mean
conf_int = forecast.conf_int()

# Visualizar as previsões até o final de 2024
plt.figure(figsize=(12, 6))
plt.plot(bitcoin_data['Close'], label='Observado')
plt.plot(pd.date_range(start=bitcoin_data.index[-1], periods=forecast_steps + 1, freq='D')[1:], forecast_values, label='Previsão', linestyle='dashed', color='orange')
plt.fill_between(pd.date_range(start=bitcoin_data.index[-1], periods=forecast_steps + 1, freq='D')[1:], conf_int.iloc[:, 0], conf_int.iloc[:, 1], color='gray', alpha=0.2, label='Intervalo de Confiança')

# Estender o gráfico fictício até 2025 (por exemplo)
extended_steps = 365  # Número de passos de extensão (um ano)
extended_dates = pd.date_range(start=bitcoin_data.index[-1] + pd.Timedelta(days=1), periods=extended_steps, freq='D')
plt.plot(extended_dates, [np.nan] * len(extended_dates), linestyle='dashed', color='orange', label='Projeção Estendida')

plt.title('Previsão do Bitcoin até 2025 com Modelo ARIMA')
plt.legend()
plt.show()
```
